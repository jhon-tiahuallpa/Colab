# Más recursos de IPython

En este capítulo, sólo hemos arañado la superficie del uso de IPython para realizar tareas de ciencia de datos.
Hay mucha más información disponible tanto en papel como en la web, y aquí enumeraremos algunos otros recursos que pueden resultarte útiles.

## Recursos en la web

- El sitio web de [IPython](http://ipython.org): El sitio web de IPython enlaza con documentación, ejemplos, tutoriales y una variedad de otros recursos.
- [El sitio web de nbviewer](http://nbviewer.jupyter.org/): Este sitio muestra representaciones estáticas de cualquier cuaderno IPython disponible en Internet. La página principal presenta algunos cuadernos de ejemplo que puedes navegar para ver para qué usa IPython otra gente.
- [Una galería de Jupyter notebooks interesantes](https://github.com/jupyter/jupyter/wiki/A-gallery-of-interesting-Jupyter-Notebooks/): Esta lista creciente de cuadernos, impulsada por nbviewer, muestra la profundidad y amplitud del análisis numérico que se puede hacer con IPython. Incluye desde breves ejemplos y tutoriales hasta completos cursos y libros compuestos en formato de cuaderno.
- Tutoriales en vídeo: buscando en Internet, encontrarás muchos tutoriales grabados en vídeo sobre IPython. Yo recomendaría especialmente buscar tutoriales de las conferencias PyCon, SciPy y PyData de Fernando Pérez y Brian Granger, dos de los principales creadores y mantenedores de IPython y Jupyter.


## Libros

- [*Python for Data Analysis*](http://shop.oreilly.com/product/0636920023784.do): El libro de Wes McKinney incluye un capítulo que cubre el uso de IPython como científico de datos. Aunque gran parte del material se solapa con lo que hemos discutido aquí, otra perspectiva es siempre útil.
- [*Learning IPython for Interactive Computing and Data Visualization*](https://www.packtpub.com/big-data-and-business-intelligence/learning-ipython-interactive-computing-and-data-visualization): Este breve libro de Cyrille Rossant ofrece una buena introducción al uso de IPython para el análisis de datos.
-  [*IPython Interactive Computing and Visualization Cookbook*](https://www.packtpub.com/big-data-and-business-intelligence/ipython-interactive-computing-and-visualization-cookbook): También de Cyrille Rossant, este libro es un tratamiento más largo y avanzado del uso de IPython para la ciencia de datos. A pesar de su nombre, no sólo trata de IPython, sino que también profundiza en una amplia gama de temas relacionados con la ciencia de datos.

Ayuda y Documentación en IPython](01.01
Ayuda-y-Documentación.ipynb)) puede ser muy útil si la usas bien y con frecuencia.
A medida que se avanza en los ejemplos aquí y en otros lugares, esto puede ser utilizado para familiarizarse con todas las herramientas que IPython tiene para ofrecer.
