# Más recursos

## Recursos de Matplotlib
Un solo capítulo de un libro nunca puede aspirar a cubrir todas las funciones y tipos de gráficos disponibles en Matplotlib.
Al igual que con otros paquetes que hemos visto, el uso liberal de las funciones de ayuda y finalización de pestañas de IPython (consulte [Ayuda y documentación en IPython](01.01-Ayuda-y-Documentación.ipynb)) puede ser muy útil al explorar la API de Matplotlib.
Además, [online documentation](http://matplotlib.org/) de Matplotlib puede ser una referencia útil.
Vea en particular el [Matplotlib gallery](http://matplotlib.org/gallery.html) vinculado en esa página: muestra miniaturas de cientos de tipos de gráficos diferentes, cada uno vinculado a una página con el fragmento de código de Python utilizado para generarlo.
De esta manera, puede inspeccionar visualmente y aprender sobre una amplia gama de diferentes estilos de trazado y técnicas de visualización.
Para un tratamiento del tamaño de un libro de Matplotlib, recomendaría [*Interactive Applications Using Matplotlib*](https://www.packtpub.com/application-development/interactive-applications-using-matplotlib), escrito por el desarrollador principal de Matplotlib, Ben Root.

## Otras bibliotecas de gráficos de Python
Aunque Matplotlib es la biblioteca de visualización de Python más destacada, también existen otras herramientas más modernas que vale la pena explorar.
Voy a mencionar algunos de ellos brevemente aquí:
- [Bokeh](http://bokeh.pydata.org) es una biblioteca de visualización de JavaScript con una interfaz de Python que crea visualizaciones altamente interactivas capaces de manejar conjuntos de datos muy grandes y/o de transmisión. El front-end de Python genera una estructura de datos JSON que puede ser interpretada por el motor Bokeh JS.
- [Plotly](http://plot.ly) es el producto de código abierto del mismo nombre de la compañía Plotly, y es similar en espíritu a Bokeh. Debido a que Plotly es el producto principal de una startup, está recibiendo un alto nivel de esfuerzo de desarrollo. El uso de la biblioteca es totalmente gratuito.
- [Vispy](http://vispy.org/) es un proyecto desarrollado activamente centrado en visualizaciones dinámicas de conjuntos de datos muy grandes. Debido a que está diseñado para apuntar a OpenGL y hacer uso de procesadores de gráficos eficientes en su computadora, puede generar algunas visualizaciones bastante grandes e impresionantes.
- [Vega](https://vega.github.io/) y [Vega](https://vega.github.io/) son representaciones gráficas declarativas y son el producto de años de investigación en el lenguaje fundamental de la visualización de datos. La implementación de representación de referencia es JavaScript, pero la API es independiente del idioma. Hay una API de Python en desarrollo en el paquete [Vega](https://vega.github.io/). Aunque a partir del verano de 2016 aún no está completamente maduro, estoy muy entusiasmado con las posibilidades de este proyecto para proporcionar un punto de referencia común para la visualización en Python y otros lenguajes.
El espacio de visualización en la comunidad de Python es muy dinámico y espero que esta lista esté desactualizada tan pronto como se publique.
